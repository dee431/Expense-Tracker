Block 2 Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside notebook
%matplotlib inline

# Better plot style
plt.style.use("ggplot")

print("Libraries Imported Successfully")
```

Block 3 Load Dataset

```python
# Replace filename if needed

df = pd.read_csv("/content/student-data-selected-columns.csv")

df.head()
```

Block 4 Basic Information

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nDataset Information:")
df.info()
```

Block 5 Check Missing Values

```python
print(df.isnull().sum())
```

Block 6 Statistical Summary

```python
df.describe()
```

Block 7 Calculate Average of Selected Column

```python
average_age = df["age"].mean()

print("Average Age:", average_age)
```

Block 8 Example:

```python
df["age"].mean()
```

Block 9 Maximum, Minimum and Median

```python
print("Maximum Age:", df["age"].max())

print("Minimum Age:", df["age"].min())

print("Median Age:", df["age"].median())
```

Block 10 Bar Chart

```python
avg = df.groupby("school")["age"].mean()

plt.figure(figsize=(8,5))

avg.plot(kind="bar", color="skyblue")

plt.title("Average Age by School")

plt.xlabel("School")

plt.ylabel("Average Age")

plt.show()
```

Block 11 Scatter Plot

```python
plt.figure(figsize=(7,5))

plt.scatter(df["age"], df["Medu"], color="red")

plt.title("Student Age vs Mother's Education")

plt.xlabel("Age")

plt.ylabel("Mother's Education (Medu)")

plt.show()
```

Block 12 Heatmap

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

Block 13 Histogram

```python
plt.figure(figsize=(7,5))

plt.hist(df["age"], bins=10)

plt.title("Distribution of Age")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

Block 14 Pie Chart (Optional)

```python
school_counts = df["school"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(school_counts,
        labels=school_counts.index,
        autopct="%1.1f%%")

plt.title("Students by School")

plt.show()
```

Block 15 Insights and Observations

```python
print("----------- INSIGHTS -----------")

print()

print("1. Average Age:", round(df["age"].mean(),2))

print("2. Highest Age:", df["age"].max())

print("3. Lowest Age:", df["age"].min())

print("4. Total Students:", len(df))

print()

print("Observations:")

print("- The heatmap shows relationships between numerical features.")

print("- The bar chart compares average age by school.")

print("- The scatter plot indicates a relationship between student age and mother's education.")

print("- The histogram shows the overall distribution of age.")
```
